```python
import os

labels_dir = "/kaggle/input/datasets/singertv/fabric-processed/processed/labels/test"
seam_images = []

for f in os.listdir(labels_dir):
    with open(os.path.join(labels_dir, f)) as file:
        lines = file.readlines()
        for line in lines:
            if line.strip().split()[0] == "4":  # seam class id
                seam_images.append(f.replace(".txt", ".jpg"))
                break

print(f"Found {len(seam_images)} seam images")
print(seam_images[:10])
```

```python
import shutil, os

src_dir = "/kaggle/input/datasets/singertv/fabric-processed/processed/images/test"
dst_dir = "/kaggle/working/seam_samples"
os.makedirs(dst_dir, exist_ok=True)

for img_name in seam_images[:5]:
    src = os.path.join(src_dir, img_name)
    dst = os.path.join(dst_dir, img_name)
    if os.path.exists(src):
        shutil.copy(src, dst)
        print(f"✅ Copied: {img_name}")
    else:
        print(f"❌ Not found: {img_name}")
```

```python
import os
from collections import Counter

labels_dirs = {
    "train": "/kaggle/input/datasets/singertv/fabric-processed/processed/labels/train",
    "val": "/kaggle/input/datasets/singertv/fabric-processed/processed/labels/val",
    "test": "/kaggle/input/datasets/singertv/fabric-processed/processed/labels/test",
}

class_names = ['Stain', 'Thread', 'Warp_Weft', 'hole', 'seam']

for split, path in labels_dirs.items():
    class_counts = Counter()
    image_counts = Counter()  # how many images contain each class at least once
    
    for fname in os.listdir(path):
        if not fname.endswith(".txt"):
            continue
        seen_classes = set()
        with open(os.path.join(path, fname)) as f:
            for line in f:
                parts = line.strip().split()
                if not parts:
                    continue
                cls_id = int(parts[0])
                class_counts[cls_id] += 1
                seen_classes.add(cls_id)
        for c in seen_classes:
            image_counts[c] += 1
    
    print(f"\n{'='*50}")
    print(f"📂 {split.upper()} SET")
    print(f"{'='*50}")
    for i, name in enumerate(class_names):
        print(f"  {name:12s} — instances: {class_counts[i]:4d}  |  images containing it: {image_counts[i]:4d}")
```
